##Make function



$f(x) = x^2 +x+1$


```
import numpy as np
import matplotlib.pyplot as plt

```

```
def f(x):
  return x**2 + x +1
```

```
# plot function


x = np.linspace(start = -3, stop = 3, num = 50)

plt.plot(x, f(x))
plt.title('$f(x) = X ^2 + X +1$', fontsize = 17)
plt.xlim(-3,3)
plt.ylim(0, 9)
plt.xlabel('x', fontsize = 14)
plt.ylabel('f(x)', fontsize = 14)


plt.show()
```

```
def df(x):
    return 2*x + 1
```

```
# Plot function and derivative

plt.figure(figsize = [14, 5])

# first : cost fuction
plt.subplot(1,2,1)
plt.plot(x, f(x))
plt.title('$f(x) = X ^2 + X +1$ (Cost fuction)', fontsize = 16, color = '#8e44ad')
plt.xlim(-3,3)
plt.ylim(0, 9)
plt.xlabel('x', fontsize = 14)
plt.ylabel('f(x)', fontsize = 14)

# 2nd : derivative fiuction
plt.subplot(1,2,2)
plt.plot(x, df(x))
plt.title('f(x) = 2x +1 (Slope of cost function)', fontsize = 16, color = "#9b59b6")
plt.xlim(-3,3)
plt.ylim(-5, 6)
plt.grid()
plt.xlabel('x', fontsize = 14)
plt.ylabel('df(x)', fontsize = 14)


plt.show()
```
