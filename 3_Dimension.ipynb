## The Curse of Dimensionality

### One dimension 

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import random
from sympy import symbols, diff

random.seed(12)
x = random.sample(range(1,101), 50)
print(x)
y = [0 for val in x]
print(y)
```

##  Visualization of 3D chart

$$f(x,y) = \frac {1}{3^{-x^2-y^2}+1}$$

make simple
$$f(x,y)=\frac{1}{r+1}$$ $r$  : $3^{-x^2-y^2}$

```
from sympy.core import numbers

def f(x,y):
    r= 3**(-x**2-y**2) 
    return  1/(r+1)

def f2(x,y):
    r= 3**(-x**2-y**2) 
    return  numbers.Float(1/(r+1))
```

```
x_4 = np.linspace(start = -2, stop = 2, num =200)
y_4 = np.linspace(start = -2, stop = 2, num =200)
# print(x_4.shape)
# print(">>>y_4", y_4)

x_4, y_4 =np.meshgrid(x_4, y_4)

```

## Partial Derivatives and Symbolic Computatoin

```
a, b = symbols("x,y")
print('Our cost function f(x,y) is :', f(a,b))
print('Partial derivaties of x is :', diff(f(a,b), a))
print('Specific value at point :', f(a,b).evalf(subs={a:1.8, b:1.0}))  # Python dictionary #key: value
print('Partial derivatives value at point :', diff(f(a,b), a).evalf(subs = {a : 1.8, b:1.0}))
```

## Batch Gradient Descent with Sympy

```
#set up
multiplier = 0.1
max_iter = 200
params = np.array([1.8, 1.0]) # initail_guess

for i in range (max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs=({a : params[0], b : params[1]}))
    gradient_y = diff(f(a, b), b).evalf(subs=({a : params[0], b : params[1]}))
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier*gradients


print('Values in gradienr array gradient :', gradients)
print('Minimum occurs at X value :' , params[0])
print('Minimum occurs at Y value :', params[1])
print('The cost is :', f(params[0], params[1]))
```

## Partial Derivatves

$$\frac{\partial f}{\partial x}=\frac{2x\ln(3)\cdot 3^{-x^2-y^2}}{(3^{-x^2-y^2}+1)^2}$$

$$\frac{\partial f}{\partial y}=\frac{2y\ln(3)\cdot 3^{-x^2-y^2}}{(3^{-x^2-y^2}+1)^2}$$

```
from math import log
from sympy.core import numbers
```

```
def fpx(x,y) :
    r = 3**(-x**2-y**2)
    return numbers.Float(2*x*log(3)*r/(r+1)**2)

```

```
#set up
multiplier = 0.1
max_iter = 200
params = np.array([1.8, 1.0]) # initail_guess

for i in range (max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier*gradients


print('Values in gradienr array gradient :', gradients)
print('Minimum occurs at X value :' , params[0])
print('Minimum occurs at Y value :', params[1])
print('The cost is :', f(params[0], params[1]))
```

```
#set up
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) # initail_guess
values_array = params.reshape(1,2)

for i in range (max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs=({a : params[0], b : params[1]}))
    gradient_y = diff(f(a, b), b).evalf(subs=({a : params[0], b : params[1]}))
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier*gradients
    values_array = np.append(values_array, params.reshape(1, 2), axis =0)


print('Values in gradient array gradient :', gradients)
print('Minimum occurs at X value :' , params[0])
print('Minimum occurs at Y value :', params[1])
print('The cost is :', f(params[0], params[1]))
```
